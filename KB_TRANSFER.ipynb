## 3. Decisiones Técnicas Clave — NO CAMBIAR

### 3.1 Infraestructura: simple_switch, NO gRPC

**Decisión**: Todos los ejercicios usan `simple_switch` (BMv2) con `simple_switch_CLI`, **NO** `simple_switch_grpc` ni P4Runtime/gRPC.

**Por qué**: La VM del curso tiene solo paquetes estándar p4lang. gRPC/P4Runtime requiere dependencias pesadas adicionales. El objetivo pedagógico es entender el plano de datos P4, no la infraestructura de control.

**Implicación práctica**:
```bash
# ✅ CORRECTO — lo que usamos
p4c-bm2-ss --p4v 16 -o p4src/build/bmv2.json p4src/main.p4
sudo python3 mininet/topo.py
simple_switch_CLI --thrift-port 9090 < s1-commands.txt

# ❌ ELIMINADO de los ejercicios originales
make                          # Makefile de p4lang (usa simple_switch_grpc)
# sX-runtime.json             # Formato P4Runtime — eliminado
# topology.json               # Descriptor run_exercise.py — eliminado
```

### 3.2 Conversión de reglas: P4Runtime JSON → simple_switch_CLI

```
# P4Runtime JSON original:
{
  "table": "MyIngress.ipv4_lpm",
  "match": { "hdr.ipv4.dstAddr": ["10.0.1.1", 32] },
  "action_name": "MyIngress.ipv4_forward",
  "action_params": { "dstAddr": "08:00:00:00:01:01", "port": 1 }
}

# simple_switch_CLI equivalente:
table_add MyIngress.ipv4_lpm MyIngress.ipv4_forward 10.0.1.1/32 => 08:00:00:00:01:01 1
# Formato: table_add <tabla> <accion> <match>[/<prefix>] => <param1> <param2>...
```

Para default actions (primera vez en MRI, tabla sin match key):
```
# P4Runtime:
{ "table": "MyEgress.swtrace", "default_action": true,
  "action_name": "MyEgress.add_swtrace", "action_params": {"swid": 1} }

# simple_switch_CLI:
table_set_default MyEgress.swtrace MyEgress.add_swtrace 1
```

- **Origen**: ONOS tutorial Exercise-3 (completado por estudiantes del curso en clase)
- **Topología**: 2 switches, 2 hosts, lineal: h1-s1-s2-h2

## 5. Diferencias Técnicas Clave Entre Ejercicios

### L2 vs L3 — Impacto en topo.py y commands.txt

| Aspectos | L2 (Ex-1,2,2.2,VLAN,MySec) | L3 (ECN, MRI) |
|----------|--------------------------|---------------|
| Tabla de forwarding | `l2_exact_table` — match exact MAC dst | `ipv4_lpm` — match LPM IP dst |
| Acción | `set_egress_port(port)` | `ipv4_forward(dstAddr_mac, port)` — reescribe src+dst MAC |
| ARP | `net.staticArp()` en topo.py | `configure_hosts()` manual con route + ARP estático |
| Subredes | Una sola (10.0.0.0/24) | Múltiples (10.0.S.0/24 por switch) |
| Gateway | No necesario | Gateway virtual (10.0.S.254 → 08:00:00:00:0S:00) |

**Patrón gateway virtual** (solo L3, en topo.py):
```python
def configure_hosts(net):
    """Configura rutas y ARP estático para que los hosts alcancen otras subredes."""
    for hname in ('h1', 'h11'):      # hosts en la subred de s1
        h = net.get(hname)
        h.cmd('route add default gw 10.0.1.254 dev eth0')
        h.cmd('arp -i eth0 -s 10.0.1.254 08:00:00:00:01:00')  # MAC del switch
    for hname in ('h2', 'h22'):      # hosts en la subred de s2
        h = net.get(hname)
        h.cmd('route add default gw 10.0.2.254 dev eth0')
        h.cmd('arp -i eth0 -s 10.0.2.254 08:00:00:00:02:00')
    # etc. para s3 en MRI
```

### Por qué send.py/receive.py son necesarios

| Ejercicio | Necesita Scapy | Razón |
|-----------|---------------|-------|
| Ex-1, 2, 2.2 | No | ping/iperf generan tráfico estándar suficiente |
| VLAN | No | Tags 802.1Q son transparentes al nivel del host |
| ECN | Semi-necesario | `send.py` envía con `tos=1` (ECN-capable); sin él el switch nunca puede marcar ECN |
| MySec | Obligatorio | IP proto 169 no existe en ninguna herramienta — Scapy lo crea desde cero |
| MRI | Obligatorio | IP Option 31 no existe; sin `send.py` el parser nunca ve MRI; sin `receive.py` los headers se muestran como hex |

### El problema del puerto en addLink() — CRÍTICO para MRI

En BMv2, el número de puerto se asigna según el orden de `addLink()` en el script Mininet. Si se reordena cualquier `addLink()`, todos los `sN-commands.txt` dejan de funcionar. El mapeo actual de MRI está documentado en `KB_MRI_ADAPTATION.md` y los comentarios de `topo.py`.
